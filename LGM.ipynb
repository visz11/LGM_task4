```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/dog.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey,cmap="gray")
```

```python
img_invert=cv2.bitwise_not(img_grey)
plt.imshow(img_invert,cmap="gray")
```

```python
img_smooth=cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
plt.imshow(img_smooth,cmap="gray")
```

```python
img_pencil=cv2.divide(img_grey,255-img_smooth,scale=255)
plt.imshow(img_pencil,cmap="gray")
```
